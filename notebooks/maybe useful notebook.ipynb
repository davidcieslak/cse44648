```python
import math, random
from matplotlib.colors import ListedColormap
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
#X as in feature vector
#Y as in class
X1 = []
X2 = []
Y  = []

#generate the circle points
for zz in xrange(1000):
    r, theta = [math.sqrt(random.random())*math.sqrt(1.0), 2*math.pi*random.random()]
    ##You can then transform r and theta back to cartesian coordinates x and y via

    x = r * math.cos(theta) 
    y = r * math.sin(theta)
    X1.append(x)
    X2.append(y)
    Y.append(0)
    
    
#generate the ring points
for zz in xrange(1000):
    r, theta = [math.sqrt(random.random())*math.sqrt(1.0) + 1.0, 2*math.pi*random.random()]
    ##You can then transform r and theta back to cartesian coordinates x and y via
    x = r * math.cos(theta)
    y = r * math.sin(theta)
    X1.append(x)
    X2.append(y)
    Y.append(1)
```

```python
X1 = np.array(X1)
X2 = np.array(X2)
Y  = np.array(Y)
```

```python
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
pl.figure()
pl.scatter(X1, X2, c=Y, cmap=cmap_bold)
pl.xlim(X1.min(), X1.max())
pl.ylim(X2.min(), X2.max())
pl.title("See? The data classes are a ring and circle!")
```
